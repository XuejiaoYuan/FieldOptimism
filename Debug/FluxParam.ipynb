```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = []
with open('flux_param.txt', 'r') as file:
    lines = file.readlines()
    for i in range(0, len(lines), 8):
        img = {}
        img['id'] = int(lines[i])
        lines[i+1] = lines[i+1].strip()
        center = lines[i+1].split()
        img['center'] = [float(center[0]), float(center[1])]
        for j in range(4):
            vertex =lines[i+2+j].strip()
            vertex = vertex.split()
            img.setdefault('vertex', []).append([float(vertex[0]), float(vertex[1])])
        
        img_x = lines[i+6].strip()
        img_x = img_x.split()
        img['x_axis'] = [float(img_x[0]), float(img_x[1])]
        img_y = lines[i+7].strip()
        img_y = img_y.split()
        img['y_axis'] = [float(img_y[0]), float(img_y[1])]
        data.append(img)
```

```python
for i in range(len(data)-300, len(data), 10):
    x = [v[0] for v in data[i]['vertex']]
    y = [v[1] for v in data[i]['vertex']]
    x.append(x[0])
    y.append(y[0])
    plt.plot(x, y)
    plt.scatter(data[i]['center'][0], data[i]['center'][1])
    plt.grid()
    plt.show()
```

```python
cos = []
for d in data:
    cos_w = np.array(d['vertex'][1]).dot(np.array([1, 0]))
    cos.append(-cos_w)
```

```python
print(len(cos))
for i in range(1000):
    plt.scatter(i, cos[i])
plt.show()
```
